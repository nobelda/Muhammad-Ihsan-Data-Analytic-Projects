```python
# create a dictionary to map the current income range values to new values
income_range_map = {
    '< 10 Juta': 'Under 10 Million',
    'Rp 10 Juta - 50 Juta': '10 Million - 50 Million',
    '> Rp 50 Juta - 100 Juta': '50 Million - 100 Million',
    '> Rp 100 Juta - 500 Juta': '100 Million - 500 Million',
    '> Rp 500 Juta - 1 Miliar': '500 Million - 1 Billion',
    '> Rp 1 Miliar': 'Over 1 Billion'
}

# use the map to replace the current income range values with the new values
df1['user_income_range'] = df1['user_income_range'].map(income_range_map)

# print the updated value counts to confirm that the changes were made successfully
print(df1['user_income_range'].value_counts())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# define the rupiah_format() function
def rupiah_format(x, pos):
    """Formats a number as Indonesian Rupiah"""
    rupiah = "Rp {:,.0f}".format(x).replace(",", ".")
    return rupiah

# create circles with radii based on the data
r = [7500000, 750000, 352500, 255000, 253993.5, 210000, 200000, 200000, 125000]

class C():
    def __init__(self,r):
        self.N = len(r)
        self.x = np.ones((self.N,3))
        self.x[:,2] = r
        maxstep = 2*self.x[:,2].max()
        length = np.ceil(np.sqrt(self.N))
        grid = np.arange(0,length*maxstep,maxstep)
        gx,gy = np.meshgrid(grid,grid)
        self.x[:,0] = gx.flatten()[:self.N]
        self.x[:,1] = gy.flatten()[:self.N]
        self.x[:,:2] = self.x[:,:2] - np.mean(self.x[:,:2], axis=0)

        self.step = self.x[:,2].min()
        self.p = lambda x,y: np.sum((x**2+y**2)**2)
        self.E = self.energy()
        self.iter = 1.

    def minimize(self):
        while self.iter < 1000*self.N:
            for i in range(self.N):
                rand = np.random.randn(2)*self.step/self.iter
                self.x[i,:2] += rand
                e = self.energy()
                if (e < self.E and self.isvalid(i)):
                    self.E = e
                    self.iter = 1.
                else:
                    self.x[i,:2] -= rand
                    self.iter += 1.

    def energy(self):
        return self.p(self.x[:,0], self.x[:,1])

    def distance(self,x1,x2):
        return np.sqrt((x1[0]-x2[0])**2+(x1[1]-x2[1])**2)-x1[2]-x2[2]

    def isvalid(self, i):
        for j in range(self.N):
            if i!=j: 
                if self.distance(self.x[i,:], self.x[j,:]) < 0:
                    return False
        return True

    def plot(self, ax):
        for i in range(self.N):
            circ = plt.Circle(self.x[i,:2],self.x[i,2], alpha=0.5, edgecolor='black', linewidth=2)
            ax.add_patch(circ)
            ax.annotate(rupiah_format(self.x[i,2], None), (self.x[i,0], self.x[i,1]), color='white', ha='center', va='center')

c = C(r)

fig, ax = plt.subplots(subplot_kw=dict(aspect="equal"))
ax.axis("off")

c.minimize()

c.plot(ax)
ax.relim()
ax.autoscale_view()
plt.title('End of Month Invested Amount by User Occupation', fontsize=16)
plt.show()
```

```python
# group the lastdate DataFrame by 'id' and count the number of transactions for each group
transaction_counts = lastdate.groupby('id')['last_transaction_date'].count()

# count how many users have each transaction frequency
frequency_counts = transaction_counts.value_counts()

# create a bar plot of transaction frequency counts
frequency_counts.plot(kind='bar', figsize=(10, 6))

# set the plot title and axis labels
plt.title('Distribution of Transaction Frequency')
plt.xlabel('Transaction Frequency')
plt.ylabel('Number of Users')

# show the plot
plt.show()
```

```python
# group the lastdate DataFrame by 'transaction_frequency' and count the number of unique 'id' values in each group
frequency_counts = lastdate.groupby('transaction_frequency')['id'].nunique()

# create a bar plot of transaction frequency by number of users
frequency_counts.plot(kind='bar', figsize=(10, 6))

# set the plot title and axis labels
plt.title('Transaction Frequency by Number of Users')
plt.xlabel('Transaction Frequency')
plt.ylabel('Number of Users')

# annotate the bars with the count of users for each frequency
for i, v in enumerate(frequency_counts):
    plt.text(i, v, str(v), color='blue', fontweight='bold', ha='center')

# show the plot
plt.show()
```

```python
# group the lastdate DataFrame by 'transaction_frequency' and count the number of unique 'id' values in each group
frequency_counts = lastdate.groupby('transaction_frequency')['id'].nunique()

# sort the frequency counts by the number of users in descending order
frequency_counts = frequency_counts.sort_values(ascending=False)

# create a bar plot of transaction frequency by number of users
frequency_counts.plot(kind='bar', figsize=(15, 6), color='tomato')

# set the plot title and axis labels
plt.title('Number of Users by Transaction Frequency')
plt.xlabel('Transaction Frequency')
plt.ylabel('Number of Users')

# annotate the bars with the count of users for each frequency
for i, v in enumerate(frequency_counts):
    plt.text(i, v, str(v), color='green', fontweight='bold', ha='center')

# show the plot
plt.show()
```

```python
# count the number of users for each transaction frequency
freq_count = lastdate['transaction_frequency'].value_counts().sort_index(ascending=False)

# calculate the cumulative percentage
cum_perc = freq_count.cumsum() / freq_count.sum()

# create a bar plot of transaction frequency by user count
ax = freq_count.plot(kind='bar', figsize=(10,6))

# create a secondary axis for the cumulative percentage
ax2 = ax.twinx()
ax2.plot(cum_perc.index, cum_perc.values, color='C1', marker='o', ms=7)

# set the plot title and axis labels
ax.set_title('Transaction Frequency by User Count')
ax.set_xlabel('Transaction Frequency')
ax.set_ylabel('User Count')
ax2.set_ylabel('Cumulative Percentage')

# add Pareto 20% line
pareto_line = 0.2 * cum_perc.max()
ax2.axhline(y=pareto_line, color='grey', linestyle='--')
ax.annotate('Pareto 20%', xy=(freq_count.index[0], pareto_line), xytext=(freq_count.index[0], pareto_line + 0.05), 
            arrowprops=dict(facecolor='black', shrink=0.05))

# sort the bars by user count from highest to lowest
ax.invert_xaxis()

# show the plot
plt.show()
```
